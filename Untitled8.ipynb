```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.ensemble import IsolationForest
from hmmlearn import hmm
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd
import numpy as np

def getData(filename):
  canData=[] #List to store CAN data
  f = open(filename)
  #read_file = reader(f)
  read_file =f.readlines()

  #file = list(read_file)
  speed = []
  rpm = []
  i = 0
  for row in read_file:
    #Change the positions of the values if needed
    record = {'stamp':row[1:18], 'PID':row[25:28], 'const1':row[29:33],
'change':row[33:41],'value':int(row[41:45], 16), 'value2':0 ,'attack':0}

    if record["PID"] == '254': #Processing of speed
      if record["value"] >= 4095:
        record["attack"] = 1
      record['value'] =  (record['value'] * 0.62137119) /100
      speed.append(record['value'])

    if record["PID"] == '115': #Processing of RPM
      if record["value"] >= 65535:
        record["attack"] = 1
      record['value'] =  (record['value'] * 2)
      rpm.append(record['value'])

    i = i+1
    canData.append(record)
    record={}

  f.close()
  return canData

def dict_to_df(dict):

  #load dictionary to dataframe
  df = pd.DataFrame.from_dict(dict)
  df = df.drop(columns=['stamp','const1','change','value2'])
  df = df.loc[(df['PID'] == '115') | (df['PID'] == '254')]
  df = df.reset_index(drop=True)
  one_hot = pd.get_dummies(df['PID'])
  df = df.drop('PID',axis = 1)
  df = df.join(one_hot)
  df = df[['115', '254', 'value', 'attack']]
  df.rename(columns = {'115':'RPM', '254':'Speed'}, inplace = True)
  df.loc[df['RPM'] == 1, 'RPM'] = df['value']
  df.loc[df['Speed'] == 1, 'Speed'] = df['value']
  df = df.drop(columns=['value'])

  return df

#edit file name with file directory of downloaded log files
fff_injection_df = dict_to_df(getData("/content/CAN Bus log - injection of FFF as the speed reading.log"))
rpm_injection_df = dict_to_df(getData("/content/CAN Bus log - injection of RPM readings.log"))
no_injection_df = dict_to_df(getData("/content/CAN bus log - no injection of messages.log"))
```

```python
fff_injection_df.replace(False, np.nan, inplace=True)
fff_injection_df.fillna(0, inplace=True)

rpm_injection_df.replace(False, np.nan, inplace=True)
rpm_injection_df.fillna(0, inplace=True)

no_injection_df.replace(False, np.nan, inplace=True)
no_injection_df.fillna(0, inplace=True)

print(fff_injection_df.head())
print(rpm_injection_df.head())
print(no_injection_df.head())
```

```python
new_fff_injection_df = fff_injection_df.drop('attack', axis=1)
new_rpm_injection_df = rpm_injection_df.drop('attack', axis=1)
new_no_injection_df = no_injection_df.drop('attack', axis=1)

print(new_fff_injection_df.head())
print(new_rpm_injection_df.head())
print(new_no_injection_df.head())
```

```python
def kmeans_clustering_with_attack(original_df, new_df, title):
    accuracies = []

    rpm_features = new_df[['RPM']]
    speed_features = new_df[['Speed']]

    # Initialize KMeans models
    kmeans_rpm = KMeans(n_clusters=2, random_state=42)
    kmeans_speed = KMeans(n_clusters=2, random_state=42)

    # Fitting the model to RPM features
    kmeans_rpm.fit(rpm_features)
    new_df['cluster_rpm'] = kmeans_rpm.labels_

    # Fitting the model to speed features
    kmeans_speed.fit(speed_features)
    new_df['cluster_speed'] = kmeans_speed.labels_

    # Centroids
    centroids_rpm = kmeans_rpm.cluster_centers_
    centroids_speed = kmeans_speed.cluster_centers_

    # Accuracy calculation
    if 'attack' in original_df.columns:
        accuracy_rpm = accuracy_score(original_df['attack'], new_df['cluster_rpm'])
        accuracies.append(accuracy_rpm)

        accuracy_speed = accuracy_score(original_df['attack'], new_df['cluster_speed'])
        accuracies.append(accuracy_speed)
    else:
        accuracies.append(None)
        accuracies.append(None)

    # Plotting RPM scatter plots
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.scatter(new_df['RPM'], np.zeros_like(new_df['RPM']), c=new_df['cluster_rpm'], cmap='viridis', marker='o', s=50, alpha=0.5)
    plt.scatter(centroids_rpm[:, 0], np.zeros_like(centroids_rpm[:, 0]), c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-Means Clustering: {title} - RPM (Accuracy: {accuracies[0] if accuracies[0] else "N/A"})')
    plt.xlabel('RPM')
    plt.yticks([])
    plt.legend()

    # Plotting Speed scatter plots
    plt.subplot(1, 2, 2)
    plt.scatter(np.zeros_like(new_df['Speed']), new_df['Speed'], c=new_df['cluster_speed'], cmap='viridis', marker='o', s=50, alpha=0.5)
    plt.scatter(np.zeros_like(centroids_speed[:, 0]), centroids_speed[:, 0], c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-Means Clustering: {title} - Speed (Accuracy: {accuracies[1] if accuracies[1] else "N/A"})')
    plt.ylabel('Speed')
    plt.xticks([])
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f'Centroids for {title} (RPM):')
    print(centroids_rpm)
    if accuracies[0]:
        print(f'Accuracy for {title} - RPM: {accuracies[0]:.2f}\n')
    else:
        print(f'Accuracy for {title} - RPM: Not available.\n')

    print(f'Centroids for {title} (Speed):')
    print(centroids_speed)
    if accuracies[1]:
        print(f'Accuracy for {title} - Speed: {accuracies[1]:.2f}\n')
    else:
        print(f'Accuracy for {title} - Speed: Not available.\n')

    return accuracies

scaler = StandardScaler()

# Scaling the original DataFrames
fff_injection_df_scaled = fff_injection_df.copy()
fff_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(fff_injection_df[['RPM', 'Speed']])

rpm_injection_df_scaled = rpm_injection_df.copy()
rpm_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(rpm_injection_df[['RPM', 'Speed']])

no_injection_df_scaled = no_injection_df.copy()
no_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(no_injection_df[['RPM', 'Speed']])

# New Dataframes
new_fff_injection_df = fff_injection_df_scaled.drop('attack', axis=1)
new_rpm_injection_df = rpm_injection_df_scaled.drop('attack', axis=1)
new_no_injection_df = no_injection_df_scaled.drop('attack', axis=1)

# calling the function
accuracy_fff = kmeans_clustering_with_attack(fff_injection_df, new_fff_injection_df, "FFF Injection")
accuracy_rpm = kmeans_clustering_with_attack(rpm_injection_df, new_rpm_injection_df, "RPM Injection")
accuracy_no_injection = kmeans_clustering_with_attack(no_injection_df, new_no_injection_df, "No Injection")
```

```python
def isolation_forest_with_attack(original_df, new_df, title):
    accuracies = []

    rpm_features = new_df[['RPM']]
    speed_features = new_df[['Speed']]

    # Initialize Isolation Forest model
    iso_forest_rpm = IsolationForest(contamination=0.1, random_state=42)
    iso_forest_speed = IsolationForest(contamination=0.1, random_state=42)

    # Fitting the model RPM
    iso_forest_rpm.fit(rpm_features)
    new_df['anomaly_rpm'] = iso_forest_rpm.predict(rpm_features)

    # Fitting the model Speed
    iso_forest_speed.fit(speed_features)
    new_df['anomaly_speed'] = iso_forest_speed.predict(speed_features)
    original_df['anomaly'] = original_df['attack'].apply(lambda x: 1 if x == 1 else -1)

    # Calculating accuracy for RPM
    accuracy_rpm = accuracy_score(original_df['anomaly'], new_df['anomaly_rpm'])
    accuracies.append(accuracy_rpm)

    # Calculating accuracy for Speed
    accuracy_speed = accuracy_score(original_df['anomaly'], new_df['anomaly_speed'])
    accuracies.append(accuracy_speed)

    # Plotting
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.scatter(new_df['RPM'], np.zeros_like(new_df['RPM']), c=new_df['anomaly_rpm'], cmap='coolwarm', marker='o', s=50, alpha=0.5)
    plt.title(f'Isolation Forest: {title} - RPM (Accuracy: {accuracy_rpm:.2f})')
    plt.xlabel('RPM')
    plt.yticks([])
    plt.colorbar(label='Anomaly (-1) / Normal (1)')

    plt.subplot(1, 2, 2)
    plt.scatter(np.zeros_like(new_df['Speed']), new_df['Speed'], c=new_df['anomaly_speed'], cmap='coolwarm', marker='o', s=50, alpha=0.5)
    plt.title(f'Isolation Forest: {title} - Speed (Accuracy: {accuracy_speed:.2f})')
    plt.ylabel('Speed')
    plt.xticks([])
    plt.colorbar(label='Anomaly (-1) / Normal (1)')

    plt.tight_layout()
    plt.show()

    print(f'Accuracy for {title} - RPM: {accuracy_rpm:.2f}')
    print(f'Accuracy for {title} - Speed: {accuracy_speed:.2f}')

    return accuracy_rpm, accuracy_speed

# Scaling the original DataFrames
scaler = StandardScaler()

# Create scaled DataFrames
fff_injection_df_scaled = fff_injection_df.copy()
fff_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(fff_injection_df[['RPM', 'Speed']])

rpm_injection_df_scaled = rpm_injection_df.copy()
rpm_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(rpm_injection_df[['RPM', 'Speed']])

no_injection_df_scaled = no_injection_df.copy()
no_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(no_injection_df[['RPM', 'Speed']])

# New Dataframes
new_fff_injection_df = fff_injection_df_scaled.drop('attack', axis=1)
new_rpm_injection_df = rpm_injection_df_scaled.drop('attack', axis=1)
new_no_injection_df = no_injection_df_scaled.drop('attack', axis=1)

# Calling the function
accuracy_fff_rpm, accuracy_fff_speed = isolation_forest_with_attack(fff_injection_df, new_fff_injection_df, "FFF Injection")
accuracy_rpm_rpm, accuracy_rpm_speed = isolation_forest_with_attack(rpm_injection_df, new_rpm_injection_df, "RPM Injection")
accuracy_no_injection_rpm, accuracy_no_injection_speed = isolation_forest_with_attack(no_injection_df, new_no_injection_df, "No Injection")
```

```python
def hmm_clustering_with_attack(original_df, new_df, title):
    accuracies = []

    # Prepare features
    rpm_features = new_df[['RPM']].values
    speed_features = new_df[['Speed']].values

    # Initialize HMM models
    model_rpm = hmm.GaussianHMM(n_components=2, covariance_type="diag", random_state=42)
    model_speed = hmm.GaussianHMM(n_components=2, covariance_type="diag", random_state=42)

    # Fit the model to RPM features
    model_rpm.fit(rpm_features)
    new_df['cluster_rpm'] = model_rpm.predict(rpm_features)

    # Fit the model to Speed features
    model_speed.fit(speed_features)
    new_df['cluster_speed'] = model_speed.predict(speed_features)

    # Accuracy calculation
    if 'attack' in original_df.columns:
        accuracy_rpm = accuracy_score(original_df['attack'], new_df['cluster_rpm'])
        accuracies.append(accuracy_rpm)

        accuracy_speed = accuracy_score(original_df['attack'], new_df['cluster_speed'])
        accuracies.append(accuracy_speed)
    else:
        accuracies.append(None)
        accuracies.append(None)

    # Plotting RPM scatter plots
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.scatter(new_df['RPM'], np.zeros_like(new_df['RPM']), c=new_df['cluster_rpm'], cmap='viridis', marker='o', s=50, alpha=0.5)
    plt.title(f'HMM Clustering: {title} - RPM (Accuracy: {accuracies[0] if accuracies[0] else "N/A"})')
    plt.xlabel('RPM')
    plt.yticks([])

    # Plotting Speed scatter plots
    plt.subplot(1, 2, 2)
    plt.scatter(np.zeros_like(new_df['Speed']), new_df['Speed'], c=new_df['cluster_speed'], cmap='viridis', marker='o', s=50, alpha=0.5)
    plt.title(f'HMM Clustering: {title} - Speed (Accuracy: {accuracies[1] if accuracies[1] else "N/A"})')
    plt.ylabel('Speed')
    plt.xticks([])

    plt.tight_layout()
    plt.show()

    print(f'Accuracy for {title} - RPM: {accuracies[0]:.2f}' if accuracies[0] is not None else f'Accuracy for {title} - RPM: Not available.')
    print(f'Accuracy for {title} - Speed: {accuracies[1]:.2f}' if accuracies[1] is not None else f'Accuracy for {title} - Speed: Not available.')

    return accuracies

# Scaling the original DataFrames
scaler = StandardScaler()

# Scale the original DataFrames
fff_injection_df_scaled = fff_injection_df.copy()
fff_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(fff_injection_df[['RPM', 'Speed']])

rpm_injection_df_scaled = rpm_injection_df.copy()
rpm_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(rpm_injection_df[['RPM', 'Speed']])

no_injection_df_scaled = no_injection_df.copy()
no_injection_df_scaled[['RPM', 'Speed']] = scaler.fit_transform(no_injection_df[['RPM', 'Speed']])

# New Dataframes
new_fff_injection_df = fff_injection_df_scaled.drop('attack', axis=1)
new_rpm_injection_df = rpm_injection_df_scaled.drop('attack', axis=1)
new_no_injection_df = no_injection_df_scaled.drop('attack', axis=1)

# Calling the function
accuracy_fff = hmm_clustering_with_attack(fff_injection_df, new_fff_injection_df, "FFF Injection")
accuracy_rpm = hmm_clustering_with_attack(rpm_injection_df, new_rpm_injection_df, "RPM Injection")
accuracy_no_injection = hmm_clustering_with_attack(no_injection_df, new_no_injection_df, "No Injection")
```
